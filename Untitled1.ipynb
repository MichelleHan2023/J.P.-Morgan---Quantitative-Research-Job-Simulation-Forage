```python
!mamba install pandas scikit-learn

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, log_loss

def process_loan_book_and_train(file_path: str) -> tuple:
    """
    Loads a full-sized credit dataset, processes variables, 
    and trains a Logistic Regression model for automated pricing.
    """
    # 1. Load the long data sheet
    # Change to pd.read_excel(file_path) if your file is an Excel file (.xlsx)
    df = pd.read_csv(file_path)
    
    # 2. Define features matching your data columns exactly
    feature_cols = [
        'credit_lines_outstanding', 
        'loan_amt_outstanding', 
        'total_debt_outstanding', 
        'income', 
        'years_employed', 
        'fico_score'
    ]
    target_col = 'default'
    
    # Clean any potential formatting issues or empty rows
    df = df.dropna(subset=feature_cols + [target_col])
    
    X = df[feature_cols]
    y = df[target_col]
    
    # 3. Stratified Split to preserve default ratios in long datasets
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.20, random_state=42, stratify=y
    )
    
    # 4. Train the predictive default model
    # Note: penalty='none' or standard C parameters keep pricing transparent for regulators
    model = LogisticRegression(max_iter=1000, random_state=42)
    model.fit(X_train, y_train)
    
    # 5. Evaluate the model performance metrics
    test_probabilities = model.predict_proba(X_test)[:, 1]
    auc_score = roc_auc_score(y_test, test_probabilities)
    cross_entropy = log_loss(y_test, test_probabilities)
    
    print("--- Model Validation Summary ---")
    print(f"Processed Dataset Size: {len(df):,} borrowers")
    print(f"Baseline Default Rate: {y.mean() * 100:.2f}%")
    print(f"Model AUC-ROC Score  : {auc_score:.4f}")
    print(f"Model Log Loss Score : {cross_entropy:.4f}\n")
    
    return model, feature_cols

def evaluate_expected_loss(borrower_data: dict, trained_model, features: list, recovery_rate: float = 0.10) -> dict:
    """
    Takes single borrower metrics and computes expected dollar losses.
    """
    # Structuring single row for inference matching features
    input_df = pd.DataFrame([borrower_data])[features]
    
    # Predict continuous Probability of Default (PD)
    pd_estimate = trained_model.predict_proba(input_df)[0, 1]
    
    # Calculate Risk exposure parameters
    exposure_at_default = borrower_data['loan_amt_outstanding']
    loss_given_default = 1.0 - recovery_rate
    
    # Quantitative Risk Equation: EL = PD * EAD * LGD
    expected_loss = pd_estimate * exposure_at_default * loss_given_default
    
    return {
        "Probability of Default (PD)": pd_estimate,
        "Exposure at Default (EAD) ($)": exposure_at_default,
        "Loss Given Default (LGD) (%)": loss_given_default * 100,
        "Expected Loss ($)": expected_loss
    }

# --- Operational Application Example ---
if __name__ == "__main__":
    # Example usage execution block:
    # 1. Provide the file path to your complete CSV/Excel sheet export
    DATA_PATH = "Task 3 and 4_Loan_Data.csv" 
    
    try:
        # 2. Train the model using the entire sheet
        trained_clf, active_features = process_loan_book_and_train(DATA_PATH)
        
        # 3. Test loss pricing calculation for a specific row/client profile
        sample_client = {
            'credit_lines_outstanding': 5,
            'loan_amt_outstanding': 4224.54,
            'total_debt_outstanding': 18858.28,
            'income': 76975.03,
            'years_employed': 7,
            'fico_score': 640
        }
        
        loss_report = evaluate_expected_loss(sample_client, trained_clf, active_features)
        
        print("--- Automated Client Risk Assessment ---")
        for metric, value in loss_report.items():
            if "$" in metric:
                print(f"{metric}: {value:,.2f}")
            elif "%" in metric:
                print(f"{metric}: {value:.1f}%")
            else:
                print(f"{metric}: {value:.4f}")
                
    except FileNotFoundError:
        print(f"Please place your full spreadsheet in the working path as '{DATA_PATH}' to execute automated training.")
```
